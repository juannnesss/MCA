```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
%matplotlib inline
```

```python
datos=np.loadtxt('monthrg.dat')
n_manchas=datos[:,3]
tiempoX=datos[:,0]+ (datos[:,1]-1)/12.0
manchas=n_manchas[n_manchas>0]
tiempo=tiempoX[n_manchas>0]
def func(t,a,w,b):
    return a*np.sin(w*t)+b
popt, pcov = curve_fit(func, tiempo, manchas)
plt.figure()
plt.scatter(tiempo,manchas)
nTiempo=np.linspace(1600,2000,2000)
fit=popt[0]*np.sin(popt[1]*nTiempo)+popt[2]
plt.plot(nTimepo,fit)
plt.show()
```
