```python
import math
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
datos=np.loadtxt('1n.dat',delimiter='&')
X=datos[:,0]
Y=datos[:,1]
plt.figure()
plt.xlabel('Corriente $\pm$ 0.1 (mA)')
plt.ylabel('Voltaje $\pm$ 0.01 (V)')
plt.scatter(X,Y)
plt.plot(X,Y)
plt.savefig('1n.png')
```
